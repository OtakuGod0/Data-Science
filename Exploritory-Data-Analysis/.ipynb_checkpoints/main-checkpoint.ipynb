# Explorator Data Analysis EDA

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df1 = pd.read_csv("Pokemon.csv", index_col = 0, encoding = 'ISO-8859-1')
```

```python
df1.head()
```

```python
df1.describe()
```

```python
df1.size
```

```python
df1.shape
```

```python
g = plt.hist('speed', histtype = 'bar', ec = 'black', bins = 40, data = df1) # bins give the number of partitions and ec give the outline col of hist
plt.xlabel('Pokemon Speed')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set_style
sns.distplot(df1['speed'])
```

```python
sns.lmplot(x = 'attack', y = 'defense', data = df1)
```

```python
sns.lmplot(x = 'attack', y = 'defense', data = df1, fit_reg = False)
```

```python
sns.lmplot(x = 'attack', y = 'defense', data = df1, fit_reg = False, hue = 'legendary')
```

```python
df1.columns
```

```python
g = sns.catplot(x = 'type1', 
                y = 'attack', 
                data = df1, 
                hue = 'legendary', 
                col = 'legendary',
                kind = 'swarm')
g.set_xticklabels(rotation = -45)
```

```python
sns.boxplot(data = df1)
```

```python
sns.jointplot(x = 'attack', y = 'defense', data = df1)
```

```python
plt.figure(figsize = (2, 2,))
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show
```

```python
t = np.arange(0., 5., 0.2)
plt.figure(figsize=(4, 3))
plt.plot(t, t, 'r-', t,t**2, 'bs', t, t*3, 'g^')
```

```python
data = {
        'a' : np.arange(50),
        'c' : np.random.randint(0,600,50),
        'd' : np.random.randn(50)
}
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d'])*100

plt.figure(figsize=(4, 3))
plt.scatter('a', 'b', c = 'c', s = 'd', data = data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
