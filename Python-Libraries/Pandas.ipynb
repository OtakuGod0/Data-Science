# Combining Datasets: Merge and Join 

Categoroes of Join
- one to one
- many to one
- many to many 

## One-to-one Join

```python
import pandas as pd 

df1 = pd.DataFrame({
    'employee': ['Bob', 'Jane', 'Lisa', 'Sue'],
    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']
})

df2 = pd.DataFrame({
    'employee': ['Lisa', 'Bob', 'Jake', 'Sue'], 
    'hire_date': [2004, 2008, 2011, 2009]
})

print(df1); print('-' * 30 ); print(df2)
```

```python
df3 = pd.merge(df1, df2)
df3
```

## Many to one join 

```python
df4 = pd.DataFrame({
    'group': ['Accounting', 'Engineering', 'HR'], 
    'supervisor': ['Carly', 'Guido', 'Steve']
})
print(df3); print("-" * 40); print(df4)
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df4, df3)
```

## Many to many Join 

```python
df5 = pd.DataFrame({
    'group': ['Accounting', 'Accounting', 'Engineering', 'Engineering', 'HR', 'HR'], 
    'skills': ['math', 'spreedsheets', 'coding', 'linux', 'spreadsheets', 'organization']
})

print(df1); print('-'*40); print(df5)
```

```python
df1.merge(df5)
```

```python
print(df1); print('-'*40);print(df2)
```

```python
df1.merge(df2, on='employee') # specify merge with keyword
```

```python
df3 = pd.DataFrame({
    'name': ['Bob', 'Jake', 'Lisa', 'Sue'], 
    "salary": [70000, 80000, 120000, 90000]
})

print(df1); print('-'*30); print(df3)
```

```python
# Joining table with no common attribute 

pd.merge(df1, df3, left_on="employee", right_on="name")
```

```python
# "name" column dropped 
pd.merge(df1, df3, left_on="employee", right_on="name").drop("name", axis = 1)
```

```python
# You can also merge on an index rather than a column

df1a = df1.set_index('employee')
df2a = df2.set_index('employee')

print(df1a); print('-'*40); print(df2a)
```

```python
df1a.merge(df2a, left_index=True, right_index=True)
```

### Join method

```python
df1a.join(df2a) # join() method performs merge that defaults to joining on indicies
```

### Specifying Set Arithmetic for Joins

```python
df6 = pd.DataFrame({
    'name': ['Peter', 'Paul', 'Mary'], 
    'food': ['fish', 'beans', 'bread']
})

df7 = pd.DataFrame({
    'name': ['Mary', 'Joseph'], 
    'drink': ['wine', 'beer']
})

print(df6); print('-'*30); print(df7)
```

```python
pd.merge(df6, df7)
```

```python
pd.merge(df6, df7, how='inner') # explicit inner join (default); return row with matching value in both dataframe
```

```python
pd.merge(df6, df7, how='outer') # fill with NaN value if no data
```

```python
# Returns all rows from the left DataFrame, and the matching rows from the right DataFrame.
pd.merge(df6, df7, how='left') 
```

```python
# Returns all rows from the right DataFrame, and the matching rows from the left DataFrame.
pd.merge(df6, df7, how='right')
```

### Overlapping Column Names: The suffixes Keyword

```python
df8 = pd.DataFrame({
    'name': ['Bob', 'Jake', 'Lisa', 'Sue'], 
    'rank': [1, 2, 3, 4]
})

df9 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'], 
                    'rank': [3, 1, 4, 2]})

print(df8); print('-'*40); print(df9)
```

```python
pd.merge(df8, df9, on="name")
```

```python
pd.merge(df8, df9, on="name", suffixes=["_left", "_right"])
```

```python
import pandas as pd 
import numpy as np

pop = pd.read_csv("state-population.csv")
areas = pd.read_csv("state-areas.csv")
abbrevs = pd.read_csv("state_names.csv").drop('Abbreviation', axis = 1)
```

```python
areas.rename(columns={'state': 'State'}, inplace = True)

print(pop.head()); print('-' * 40); print(areas.head());print('-'*40);print(abbrevs.head())
```

```python
merged = pd.merge(pop, abbrevs, left_on='state/region', right_on='Alpha code', how='left')
merged.drop(['Code', 'Alpha code'], axis = 1, inplace = True)
merged.head()
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()].head()
```

```python
merged[merged['State'].isnull()]['state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'State'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'State'] = 'United States'
merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on = 'State', how = 'left')
final.head()
```

```python
final.isnull().any()
```

```python
final['State'][final['population'].isnull()].unique()
```

```python
final.dropna(inplace = True)
final.isnull().any()
```

```python
# query method 
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()
```
