```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("president_heights.csv")
df = df.sort_values(by = "height(cm)")
combined_data = pd.concat([df.head(5), df.tail(5)])
plt.plot(combined_data["name"], combined_data["height(cm)"], marker = "o")
plt.xticks(rotation=90)
```
