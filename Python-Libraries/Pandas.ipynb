## By Dictionary 

```python
population_dict = {'california': 32, 
                   'texas': 26, 
                   'New York': 20, 
                   'florida': 19, 
                   'illinois': 13
                  }
population = pd.Series(population_dict)
population
```

```python
population['california':'New York'] # series is inclusive here meaning New york is also included 
```

```python
pd.Series(7,  index=[10, 20, 30, 40]) # when only one value is passed for multiple index
```

```python
pd.Series({2:'a', 5:'b', 10:'c'}, index = [5, 2]) # overwriting dict with index parameter
```
