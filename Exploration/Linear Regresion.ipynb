# Advertising dataset 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set()

df = pd.read_csv("advertising.csv")
df.columns
```

## Visualizing data

```python
df["Sum"] = df["TV"] + df["Radio"] + df["Newspaper"]
plt.scatter(df["TV"], df["Sales"], color="blue", label = "TV")
plt.scatter(df["Radio"], df["Sales"], color="green", label = "Radio")
plt.scatter(df["Newspaper"], df["Sales"], color="red", label = "Newspaper")
plt.scatter(df["Sum"], df["Sales"], color="violet", label = "Sum")
plt.ylabel("Sales")
plt.legend(title = "Advertising")
```

## Training model 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# spliting data to train and test
x = df["TV"].values.reshape(-1, 1)
y = df["Sales"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3)

# training model 
lr = LinearRegression()
lr.fit(x_train, y_train)

# visualizing data 
plt.scatter(x_test, y_test, color = "blue")
plt.plot(x_train, lr.predict(x_train), color = "red")
```

## Calculating error 

```python
from sklearn import metrics

y_pred = lr.predict(x_test)
MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = metrics.root_mean_squared_error(y_test, y_pred)

print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
```

## Correlation between Sum and Advertisment

```python

# Splitting data into train and test sets
sum = (df["TV"] + df["Radio"] + df["Newspaper"]).values.reshape(-1, 1)  # Convert to 2D array
y = df["Sales"]
x_train, x_test, y_train, y_test = train_test_split(sum, y, test_size=1/3)

# Training the model
lr = LinearRegression()
lr.fit(x_train, y_train)

# Visualizing test data
plt.scatter(x_test, y_test, color="blue")
plt.plot(x_test, lr.predict(x_test), color="red")

# Show plot
plt.xlabel("Total Spend on Advertising")
plt.ylabel("Sales")
plt.title("Sales vs Total Advertising Spend")
plt.show()

# Calculating error
print("MAE:", metrics.mean_absolute_error(y_test, lr.predict(x_test)))
print("MSE:", metrics.mean_squared_error(y_test, lr.predict(x_test)))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, lr.predict(x_test))))
```
